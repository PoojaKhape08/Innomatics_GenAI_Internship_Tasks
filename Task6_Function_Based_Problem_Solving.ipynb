Problem Statement 1:Smart Parking Lot Management System

Design a function to manage a smart parking lot.

The system should:
1. Accept vehicle entry and exit logs
2. Calculate total parked vehicles
3. Identify peak parking usage
4. Alert if parking exceeds capacity

```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked = parked + 1
        elif log == "OUT" and parked > 0:
            parked = parked - 1

        if parked > peak:
            peak = parked

    if parked > capacity:
        status = "Exceeded Capacity"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Parking Status:", status)

parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

Problem Statement 2:Online Food Delivery Time Estimator

Create a function that estimates delivery time based on:
1. Distance (km)
2. Weather condition
3. Traffic level

```python
def delivery_time(distance, traffic, weather):
    time = distance * 5  # base time

    if traffic == "High":
        time=time+15
    elif traffic == "Medium":
        time=time+ 8

    if weather == "Rainy":
        time=time+10

    print("Estimated Delivery Time:", time, "minutes")

delivery_time(8, "High", "Rainy")
```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

Build a function that analyzes seat booking data and:
1. Calculates occupancy percentage
2. Determines if show is Housefull
3. Suggests opening additional shows

```python
def theatre_occupancy(total_seats, booked):
    occupancy = (len(booked) / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)

theatre_occupancy(200, [1]*150)
```

Problem Statement 4:Cloud Server Load Classification System

Create a function to classify server load based on CPU usage readings.

Rules:
1. Average CPU < 50% → Normal
2. 50%–80% → Warning
3. 80% → Critical


```python
def server_load(cpu):
    avg = sum(cpu) / len(cpu)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)

server_load([45, 60, 70, 85, 90])
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor
 
Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.

```python
def classroom_monitor(resources):
    overused = []

    for r, hours in resources.items():
        if hours > 8:
            overused.append(r)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)

classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

 Problem Statement 6: Online Event Registration Capacity Controller

Create a function that manages event registrations by:
1. Tracking registrations
2. Preventing overbooking
3. Triggering waitlist mode

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

event_registration(100, 105)
```
