## Task 1: Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as two different words because computers are case-sensitive.
This can increase vocabulary size unnecessarily and affect model performance.
To solve this, text is usually converted into lowercase so that both words are treated the same.

2. What happens if stopwords are not removed?


1) The dataset becomes noisy and large
2) Model processing becomes slower
3) Important words get less importance
4) Accuracy of the model may decrease

3. Provide two real-world scenarios where removing stopwords can be harmful.

1) In sentiment analysis:
Example: "I am not happy"
Removing "not" changes the meaning completely.
2) In question answering systems:
Example: "What is your name?"
Removing stopwords may remove important context words like "what" or "is".

4. Explain the difference between stemming and lemmatization.

Stemming:
Cuts words to their root form without considering meaning
Example: "loving" → "lov"

Lemmatization:
Converts words into their meaningful base form using vocabulary
Example: "loving" → "love"
Lemmatization is more accurate than stemming.

## Task 2: Advanced Preprocessing Function

```python
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)
    
    # 3. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 4. Remove repeated characters
    text = re.sub(r'(.)\1+', r'\1\1', text)
    
    # 5. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 6. Remove extra spaces
    text = " ".join(text.split())
    
    # 7. Tokenization
    tokens = text.split()
    
    # 8. Remove short words
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    # 9. Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    # 10. Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Final cleaned sentence
    cleaned_text = " ".join(tokens)
    
    return tokens, cleaned_text
```

## Task 3: Stress Testing

```python
test_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in test_data:
    tokens, cleaned = preprocess_text(sentence)
    
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-"*50)
```

## Task 4: Token Analytics

```python
for sentence in test_data:
    tokens, _ = preprocess_text(sentence)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0
    
    print("Sentence:", sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-"*50)
```

### Analysis:

1. Which sentence had the most noise?

The sentence "Win $$$ now!!! Limited offer!!!" had the most noise due to special characters, repeated symbols, and unnecessary formatting.

2. Which sentence retained the most meaningful tokens after cleaning?

The sentence "I absolutely looooved this product 😍😍" retained meaningful tokens like "absolutely", "loved", and "product" after preprocessing.

## Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in test_data:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

## Task 6: Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_data)

print("All Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```
