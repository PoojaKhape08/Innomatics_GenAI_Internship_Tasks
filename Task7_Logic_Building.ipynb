Problem 1: Smart Light Controller Problem Statement Create a class to control a smart light that can be turned ON or OFF and display its status. 

```python
class SmartLight:    
    def __init__(self, light_name):   # Correct constructor
        self.light_name = light_name
        self.status = "OFF"    
    def turn_on(self):
        self.status = "ON"
        print(self.light_name, "is", self.status)    
    def turn_off(self):
        self.status = "OFF"
        print(self.light_name, "is", self.status)    
    def display_status(self):
        print(self.light_name, "is", self.status)
light = SmartLight("Bedroom Light")
light.turn_on()
```

Problem 2: Employee ID Card System Problem Statement Create a class that stores employee name, ID, and department and displays ID card details.

```python
class Employee:    
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department    
    def display_id_card(self):
        print("Employee ID Card")
        print("Employee Name:", self.name)
        print("Employee ID:", self.emp_id)
        print("Department:", self.department)
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

Problem 3: Mobile Contact Record Problem Statement Create a class to store contact name and phone number and display contact information.

```python
class Contact:    
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone    
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
contact1 = Contact("Anita", "9876543210")
contact1.display_contact()
```

Problem 4: Product Price Tag Generator Problem Statement Create a class that stores product name and price and prints a formatted price tag.

```python
class Product:    
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))
item = Product("Headphones", 2499)
item.print_price_tag()
```

Problem 5: Movie Rating Display System Problem Statement Create a class that stores movie name and rating and displays movie details.

```python
class Movie:    
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating    
    def display_details(self):
        print("Movie Name:", self.name)
        print("Rating:", self.rating)
movie1 = Movie("Inception", 4.8)
movie1.display_details()
```

Problem 6: Delivery Address Manager Problem Statement Create a class that stores customer name and delivery address and prints delivery details. 

```python
class Delivery:    
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address    
    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
order1 = Delivery("Suman", "Hyderabad")
order1.print_delivery_details()
```
