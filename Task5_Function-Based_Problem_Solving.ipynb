Problem 1: Social Media – Post Engagement Analyzer

Problem Statement:-  
Create a function to analyze engagement on a social media post.
    
Rules  
1. Calculate total likes
2. If Total Likes ≥ 1000 → Viral Post
3. Otherwise → Normal Engagement

```python
def post_engagement(likes):
    total = 0
    for l in likes:
        total = total + l

    if total >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total)
    print("Post Status:", status)

post_engagement([200, 300, 250, 300])
```

Problem 3: Agriculture – Rainfall Adequacy Checker

Problem Statement:-  
Create a function that analyzes seasonal rainfall data.

```python
def rainfall_check(data, required):
    total = 0
    for r in data:
        total = total + r

    avg = total / len(data)

    if avg >= required:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(avg))
    print("Rainfall Status:", status)

rainfall_check([70, 75, 72, 71], 70)
```

Problem 4: Social Media – Duplicate Account Detection

Problem Statement:-    
Create a function to detect duplicate usernames in a system.

Rules
1. If duplicates exist → Flag accounts
2. Else → Accounts are unique

```python
def duplicate_accounts(users):
    if len(users) != len(set(users)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["user1","user2","user1"])
```

Problem 6: Agriculture – Premium Crop Price Filter

Problem Statement:-   
Create a function that filters crop prices above a premium threshold.

Rules
1. Price > ₹2000 → Premium Crop
2. Else → Regular Crop

```python
def premium_crops(prices):
    premium = []

    for p in prices:
        if p > 2000:
            premium.append(p)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

Problem 9: E-Learning – Student Attendance Eligibility System

Problem Statement:-       
Create a function to calculate attendance percentage and check exam eligibility.
    
Rules
1. Attendance ≥ 75% → Eligible
2. Attendance < 75% → Not Eligible

```python
def attendance_check(att):
    present = 0

    for a in att:
        if a == "P":
            present = present + 1

    percent = (present / len(att)) * 100

    if percent >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percent)
    print("Exam Eligibility:", status)

attendance_check(["P","P","A","P","P"])
```

Problem 10: Smart Electricity Bill Analyzer

Problem Statement:-    
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
Billing

Rules
1. First 100 units → ₹3 per unit
2. Next 100 units (101–200) → ₹5 per unit
3. Above 200 units → ₹7 per unit

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100*3) + (units-100)*5
    else:
        bill = (100*3) + (100*5) + (units-200)*7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    print("Total Bill:", bill)
    print("Usage Level:", usage)

electricity_bill(250)
```
