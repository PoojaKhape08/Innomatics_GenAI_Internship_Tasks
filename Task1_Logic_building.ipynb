1.User Login Check

Problem Statement:
Given a username and password, check whether login is successful.

username = "admin"

password = "1234"

Requirements:
Print "Login Successful" if both username and password match Otherwise print "Invalid Credentials"


```python
username = "admin"
password = "1234"

user = input("Enter username: ")
passw = input("enter password: ")

if user == username and passw == password:
    print("Login successful")
else:
    print("Invalid Credentials")
```

2.Pass / Fail Analyzer

Problem Statement:
Analyze student results.

marks = [45, 78, 90, 33, 60]


Requirements:
A student passes if marks â‰¥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly


```python
marks = [45, 78, 90, 33, 60]

pass_count=0
fail_count=0

for m in marks:
    if m>=50:
        pass_count= pass_count + 1
    else:
        fail_count= fail_count + 1

print("Pass Students: ",pass_count)
print("Fail Student: ",fail_count)
```

3.Simple Data Cleaner

Problem Statement:
Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]

Requirements:
Remove extra spaces
Convert all names to lowercase
Print the cleaned list


```python
names = ["Alice","bob","CHARLIE"]
cleaned_names = []

for name in names: 
    name = name.strip()     #remove spaces
    name = name.lower()     #convert to lowercase
    cleaned_names.append(name)
print(cleaned_names)
```

4.Message Length Analyzer

Problem Statement:
Analyze message sizes.
    
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters


```python
messages = ["Hi","Welcome to the platform","OK"]

for msg in messages:
    print(msg,"length: ", len(msg))

    if len(msg)>10:
        print("Long message detected")
```

5.Error Message Detector

Problem Statement:
Detect error messages from system logs.
    
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count=0

for log in logs:
    if log == "ERROR":
        error_count=error_count + 1
print("Total Errors: ",error_count)
```
