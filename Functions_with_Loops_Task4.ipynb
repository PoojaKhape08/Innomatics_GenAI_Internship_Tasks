Problem 1: Mobile Recharge Validation System

Problem Statement:

Create a function to validate a mobile recharge request.

Rules:
1. Recharge amount must be ≥ ₹50
2. Valid plans: 199, 299, 399, 599
3. User should be allowed to retry if invalid


```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge amount must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("Invalid plan. Choose from 199, 299, 399, 599")
        else:
            print("Recharge Successful!")
            break

validate_recharge()
```

Problem 2: Inventory Reorder Alert System

Problem Statement:

Given a dictionary of products and stock quantities:
1. Print Reorder Alert if stock < 15
2. Print Stock OK otherwise

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, "→ Reorder Alert")
        else:
            print(product, "→ Stock OK")

products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Soap": 25
}

check_inventory(products)
```

Problem 3: Student Result Processing System

Problem Statement:-

Create a function that:
1. Accepts a list of student marks
2. Calculates average marks
3. Prints Pass if average ≥ 50 else Fail

```python
def check_result(marks):
    total = 0

    for m in marks:
        total = total + m

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 55, 70, 40]
result = check_result(marks)

print("Average Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

Problem Statement:-
    
Create a cab fare calculator:
1. Base fare = ₹50
2. ₹12 per km
3. Peak hour → 25% extra
4. Allow retry using while loop

```python
def calculate_fare():
    while True:
        km = float(input("Enter distance in km: "))
        peak = input("Peak hour? (yes/no): ")

        fare = 50 + (12 * km)

        if peak.lower() == "yes":
            fare = fare + fare * 0.25

        print("Total Fare:", fare)

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break

calculate_fare()
```

Problem 5: Employee Attendance Eligibility Checker

Problem Statement:-

Create a function that:
1. Accepts attendance list ("P", "A")
2. Calculates attendance percentage
3. Returns Eligible if ≥ 75% else Not Eligible

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present = present + 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

Problem 6: Password Strength Checker

Problem Statement:-
    
Create a function to check password strength.
    
Rules:
1. Minimum length = 8
2. Must contain at least one digit
3. Must contain at least one special character (@#$)

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
